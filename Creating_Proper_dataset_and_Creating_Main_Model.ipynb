```python
import os
from ultralytics import YOLO
import cv2
import shutil
```

```python
def delete_previous_labels(folder_path):
  for file in os.listdir(folder_path):
    if(file.split('.')[1]=='txt'):
      os.remove(folder_path+"/"+file)
      print("Successfully deleted : ",file)
```

```python
shutil.copytree('/content/drive/MyDrive/Cancer_Cell_Detection','/content/Cancer_Cell_Detection')
```

```python
Images_Train_dir='/content/Cancer_Cell_Detection/TCIA_SegPC_dataset/train/y'
Images_Train_labels_paths='/content/Cancer_Cell_Detection/TCIA_SegPC_dataset/train/x'
Images_Validation_dir='/content/Cancer_Cell_Detection/TCIA_SegPC_dataset/validation/y'
Images_Validation_labels_paths='/content/Cancer_Cell_Detection/TCIA_SegPC_dataset/validation/x'
```

```python
delete_previous_labels(Images_Train_dir)
delete_previous_labels(Images_Train_labels_paths)
delete_previous_labels(Images_Validation_dir)
delete_previous_labels(Images_Validation_labels_paths)
```

```python
shutil.copyfile('/content/Cancer_Cell_Detection/classes.txt',Images_Train_labels_paths+"/classes.txt")
shutil.copyfile('/content/Cancer_Cell_Detection/classes.txt',Images_Validation_labels_paths+"/classes.txt")
```

```python
model=YOLO('/content/Cancer_Cell_Detection/Temp_weights/best.pt')
```

```python
def Generate_labels(Images_dir):
  for file in os.listdir(Images_dir):
    img=cv2.imread(Images_dir+"/"+file)
    height=img.shape[0]
    width=img.shape[1]
    try:
        x1,y1,x2,y2=model.predict(Images_dir+"/"+file)[0].boxes.xyxy[0]
        x1=float(x1)
        y1=float(y1)
        x2=float(x2)
        y2=float(y2)
        l=abs(x2-x1)
        b=abs(y2-y1)
        xmin=min(x1,x1)
        ymin=min(y1,y1)
        data="0 "+str(float((xmin+(l/2))/width))+" "+str(float((ymin+(b/2))/height))+" "+str(float(l/width))+" "+str(float(b/height))
        text_file = open(os.getcwd()+"/labels_Folders/"+str(file.split(".")[0])+".txt", "w+")
        text_file.write(data)
        text_file.close()
        print("Done")
    except:
        pass
def merging_Files(Output_path):
  files=os.listdir(os.getcwd()+"/labels_Folders")
  for single_file in files:
    if(single_file.split("_")[0]+".txt" in os.listdir(Output_path)):
      file1 = open(Output_path+"/"+single_file.split("_")[0]+".txt", "a")
      file2=open(os.getcwd()+'/labels_Folders/'+single_file,"r")
      data=file2.read()
      file1.write(data+"\n")
      file1.close()
      file2.close()
      print("Successfully appended to ",single_file.split("_")[0]+".txt")
    else:
      file1 = open(Output_path+"/"+single_file.split("_")[0]+".txt", "w+")
      file2=open(os.getcwd()+'/labels_Folders/'+single_file,"r")
      data=file2.read()
      file1.write(data+"\n")
      file1.close()
      file2.close()
      print("Successfully Created a new file and appended to ",single_file.split("_")[0]+".txt")
```

```python
os.mkdir(os.getcwd()+"/labels_Folders")
Generate_labels(Images_Train_dir)
merging_Files(Images_Train_labels_paths)
shutil.rmtree(os.getcwd()+"/labels_Folders")
```

```python
os.mkdir(os.getcwd()+"/labels_Folders")
Generate_labels(Images_Validation_dir)
merging_Files(Images_Validation_labels_paths)
shutil.rmtree(os.getcwd()+"/labels_Folders")
```
